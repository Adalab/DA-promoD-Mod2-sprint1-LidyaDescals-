3. Preparación de los datos:
- Los nombres de las columnas empiezan con mayúsculas, poned todos los nombres de las columnas en minúsculas.


```python
df.rename(columns = {col:col.lower() for col in df.columns}, inplace = True)
```

- Haciendo el análisis exploratorio nos deberíamos haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que deberían. Esto es debido a que los decimales están establecidos como comas y no con puntos. Cread una función que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.


```python
def coma_por_punto(elem):
    '''
    Reemplaza las comas por puntos en el parámetro que le pasemos.
    Args: 
        elem (str): cada valor de cada columna a la que le aplicamos la función.
    Returns:
        El valor modificado en formato string. Si no puede hacer el reemplazo, devuelve el valor sin modificar.
    '''
    try:
        return elem.replace(',','.')
    except:
        return elem

columnas_a_float = ["danceability", "energy", "key", "loudness", "speechiness", "acousticness", "instrumentalness"]

for col in columnas_a_float:
    df[col] = df[col].apply(coma_por_punto)

```

- Después de haber hecho los cambios, chequead los tipos de datos. ¿Son ya del tipo correcto? En caso de que no, cread otra función o modificar la anterior para que sean de tipo float.


```python
df.dtypes
```

```python
dicc = {i: float for i in columnas_a_float}

df = df.astype(dicc)
```

```python
df.dtypes
```

- Hay algunas canciones cuyo título está en mayúsulas. Crea una función para que todos los títulos estén en minúscula. Haced lo mismo para las columnas de "track" y "album".


```python
def minusculas(elem):
    '''
    Toma un elemento y cambia todas sus letras a minúsculas.
    Args: 
        elem (str): el string al que queramos aplicarle la función.
    Returns:
        El elemento en formato string y en minúsculas.
    '''
    try:
        return elem.lower()
    except:
        return elem

columnas_a_minus = ["title", "track", "album"]

for col in columnas_a_minus:
    df[col] = df[col].apply(minusculas)   
```

- Algunos de los artistas tienen símbolos raros, en concreto el símbolo $. Cread una función que nos elimine este símbolo de los nombres de los artistas.


```python
def eliminar_simbolo(elem):
    '''
    Elimina el símbolo $ de un string.
    Args: 
        elem (str): el string al que queramos aplicarle la función.
    Returns:
        El elemento en formato string y sin el símbolo $.
    '''
    try:
        return elem.replace('$','')
    except:
        return elem

df["artist"] = df["artist"].apply(eliminar_simbolo)
```

- Se quiere estudiar si existe una relación entre la cantidad de visitas que reciben los videos de música de un artista en una plataforma de streaming y la cantidad de likes que estos videos obtienen. Por lo tanto, querremos contestar a la siguiente pregunta: ¿Son aquellos artistas que tienen más visitas los que más likes tienen? Para solucionar este ejercicio deberéis:
    - Agrupar por artista y calcular la media de visitas y de likes.


```python
visitas_likes = df.groupby("artist")[["views", "likes"]].mean()
visitas_likes
```

- Quedaros con los 10 artistas que más visitas han tenido.


```python
visitas_likes_v = visitas_likes.sort_values(by = "views", ascending = False).iloc[:10]
visitas_likes_v
```

- Quedaros con los 10 artistas que más likes han tenido.


```python
visitas_likes_l = visitas_likes.sort_values(by = "likes", ascending = False).iloc[:10]
visitas_likes_l
```

- Haced un subplot, donde en uno tengamos los 10 artistas con más likes y en el otro los 10 artistas con mas views.


```python
visitas_likes_l = visitas_likes_l.reset_index().sort_values(by="likes", ascending= True)
visitas_likes_v = visitas_likes_v.reset_index().sort_values(by="views", ascending= True)
my_range=range(0,len(visitas_likes_l.index))
```

```python
fig,axes = plt.subplots(1, 2, figsize = (15, 5))
fig.suptitle('Likes y visitas en Youtube por artista', fontsize = 20)
#Por Visitas
axes[0].hlines(y=visitas_likes_v['artist'], xmin = 0, xmax=visitas_likes_v['views'], color='orange')
axes[0].plot(visitas_likes_v['views'], my_range, "o", markersize = 12)
axes[0].set_title("Los diez artistas con mayor número de visitas en Youtube", loc='left')
axes[0].set_xlabel('Número de visitas (en millones)')
axes[0].set_ylabel('Artista')
axes[0].spines[['right','top']].set_visible(False)

#Por Likes
axes[1].hlines(y=visitas_likes_l['artist'], xmin = 0, xmax=visitas_likes_l['likes'], color='orange')
axes[1].plot(visitas_likes_l['likes'], my_range, "o", markersize = 12)
axes[1].set_title("Los diez artistas con mayor número de likes en Youtube", loc='right')
axes[1].set_xlabel('Número de likes (en millones)')
axes[1].set_ylabel('Artista')
axes[1].spines[['right','top']].set_visible(False)
plt.tight_layout()
plt.show()
```
